**Perform Feature Scaling**

**1. Using MinMaxScaler (Built-in Library)
Min-Max Scaling transforms data to a range of [0, 1]**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
df['Age'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
```

**2. Standard Scaling (Z-Score Normalization)
Standard Scaling transforms features to have a mean of 0 and a standard deviation of 1.**

(a) Using Built-in StandardScaler

```python
from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()
```

```python
df[['Age']] = s_scaler.fit_transform(df[[ 'Age']])
```

```python
df['Age']
```

```python
df['Age'] = (df['Age'] - df['Age'].mean()) / df['Age'].std()
```
